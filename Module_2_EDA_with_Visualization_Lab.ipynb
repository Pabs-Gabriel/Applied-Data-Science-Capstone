```python
import pandas as pd
import requests

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"

# Download the CSV and load it into a DataFrame
df = pd.read_csv(url)

# Show the first 5 rows
df.head()
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
# Task 1 Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    data=df,
    x="FlightNumber",
    y="LaunchSite",
    hue="Class",
    kind="strip",
    height=6,
    aspect=1.5
)

plt.title("Flight Number vs Launch Site (colored by Class)")
plt.show()
```

```python
#Task 2 # Plot a scatter point chart with x axis to be Pay Load Mass (kg) and y axis to be the launch site, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot: Payload Mass vs Launch Site, hue by Class
sns.catplot(data=df, x="PayloadMass", y="LaunchSite", hue="Class", kind="strip", height=6, aspect=1.5)

plt.title("Payload Mass vs Launch Site (colored by Class)")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Launch Site")
plt.show()
```

```python
#Task 3 Visualize the relationship between success rate of each orbit type
# HINT use groupby method on Orbit column and get the mean of Class column

import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Group by Orbit and get mean of Class (success rate)
orbit_success = df.groupby("Orbit")["Class"].mean().reset_index()

# Step 2: Plot as bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=orbit_success, x="Orbit", y="Class", palette="Blues_d")

plt.title("Success Rate by Orbit Type")
plt.ylabel("Success Rate")
plt.xlabel("Orbit Type")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```

```python
# TASK  4: Visualize the relationship between FlightNumber and Orbit type
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    data=df,
    x="FlightNumber",
    y="Orbit",
    hue="Class",
    kind="strip",
    height=6,
    aspect=1.5
)

plt.title("Flight Number vs Orbit Type (colored by Class)")
plt.xlabel("Flight Number")
plt.ylabel("Orbit")
plt.show()
```

```python
#TASK 5: Visualize the relationship between Payload Mass and Orbit type
# Plot a scatter point chart with x axis to be Payload Mass and y axis to be the Orbit, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    data=df,
    x="PayloadMass",   # Adjust this if your column is named differently
    y="Orbit",
    hue="Class",
    kind="strip",
    height=6,
    aspect=1.5
)

plt.title("Payload Mass vs Orbit Type (colored by Class)")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Orbit")
plt.show()
```

```python
#TASK 6: Visualize the launch success yearly trend
# A function to Extract years from the date
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
# Step 1: Extract year from "Date" and overwrite the column or store in new column
df['Year'] = df["Date"].str.split("-").str[0]

# Step 2: Group by year and calculate success rate (mean of Class column)
yearly_success = df.groupby("Year")["Class"].mean().reset_index()

# Step 3: Plot the trend
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_success, x="Year", y="Class", marker="o")

plt.title("Yearly Launch Success Trend")
plt.xlabel("Year")
plt.ylabel("Average Success Rate")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
