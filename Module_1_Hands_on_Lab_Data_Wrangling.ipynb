```python
# Apply value_counts on Orbit column
import pandas as pd

# Load the SpaceX dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv")

# Use value_counts() to count number of launches for each site
orbit_counts = df['Orbit'].value_counts()

# Display the results
print(orbit_counts)
```

```python
# Count the number of occurrences of each unique landing outcome
landing_outcomes = df['Outcome'].value_counts()

# Display the result
print(landing_outcomes)
```

```python
for i,outcome in enumerate(landing_outcomes.keys()):
    print(i,outcome)
```

```python
bad_outcomes=set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

```python
# Define bad outcomes
bad_outcome = set(['None', 'Failure', 'False Ocean', 'False ASDS', 'False RTLS', 'Precluded'])

# Create landing_class list using list comprehension
landing_class = [0 if outcome in bad_outcome else 1 for outcome in df['Outcome']]

# Optional: Add it as a column to the dataframe
df['Class'] = landing_class

# Display first few values to verify
print(df[['Outcome', 'Class']].head())
```

```python
df['Class']=landing_class
df[['Class']].head(8)
```

```python
# Filter the dataframe for CCAFS SLC 40
ccafs_df = df[df['LaunchSite'] == 'CCAFS SLC 40']

# Display outcome counts
print(ccafs_df['Outcome'].value_counts())

# Calculate success rate
successes = ccafs_df['Outcome'].str.contains('Success').sum()
total = len(ccafs_df)
success_rate = (successes / total) * 100

print(f"Success rate at CCAFS SLC 40: {success_rate:.2f}%")
```

```python
# Count how many launches went to geosynchronous orbit (GTO)
gto_count = df['Orbit'].value_counts().get('GTO', 0)
print("GTO Launches:", gto_count)
```

```python
# Count successful drone ship landings
drone_success_count = df['Outcome'].value_counts().get('Success (drone ship)', 0)
print("Successful drone ship landings:", drone_success_count)
```
