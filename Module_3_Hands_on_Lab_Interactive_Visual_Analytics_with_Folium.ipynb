```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Provided coordinates of proximities
coast_coord = [28.563197, -80.576820]   # Near CCAFS
rail_coord = [28.572872, -80.585835]
highway_coord = [28.56342, -80.5708]
city_coord = [28.3922, -80.6077]        # Cocoa Beach

# Function to calculate distance
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# Add distance columns to launch_sites DataFrame
launch_sites['Distance_to_Coast_km'] = launch_sites.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], *coast_coord), axis=1)

launch_sites['Distance_to_Rail_km'] = launch_sites.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], *rail_coord), axis=1)

launch_sites['Distance_to_Highway_km'] = launch_sites.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], *highway_coord), axis=1)

launch_sites['Distance_to_City_km'] = launch_sites.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], *city_coord), axis=1)

# Display result
launch_sites[['Launch Site', 'Distance_to_Coast_km', 'Distance_to_Rail_km',
              'Distance_to_Highway_km', 'Distance_to_City_km']]
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
import numpy as np

# Distance calculation function (already defined)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# Launch site (CCAFS LC-40)
launch_name = "CCAFS LC-40"
launch_lat, launch_lon = 28.562302, -80.577356

# Proximities (replace with your MousePosition results if needed)
coast = [28.56367, -80.57163]
highway = [28.56342, -80.5708]
railway = [28.572872, -80.585835]
city = [28.3922, -80.6077]  # Cocoa Beach

# Dictionary of places
places = {
    "Coastline": coast,
    "Highway": highway,
    "Railway": railway,
    "Nearest City": city
}

# Initialize map
site_map = folium.Map(location=[launch_lat, launch_lon], zoom_start=12)

# Add launch site marker
folium.Marker(
    [launch_lat, launch_lon],
    popup=launch_name,
    icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
).add_to(site_map)

# Draw distances and lines
for label, coord in places.items():
    # Distance
    distance = calculate_distance(launch_lat, launch_lon, coord[0], coord[1])

    # Add marker at the point
    folium.Marker(
        location=coord,
        popup=f"{label}: {distance:.2f} km",
        icon=folium.Icon(color='green', icon='info-sign')
    ).add_to(site_map)

    # Add line between launch site and point
    folium.PolyLine(
        locations=[[launch_lat, launch_lon], coord],
        color='red',
        weight=2
    ).add_to(site_map)

    # Optional: Add text label in the middle of the line
    mid_lat = (launch_lat + coord[0]) / 2
    mid_lon = (launch_lon + coord[1]) / 2
    folium.Marker(
        location=[mid_lat, mid_lon],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 10pt; color: darkred;"><b>{distance:.2f} km</b></div>',
        )
    ).add_to(site_map)

# Display the map
site_map
```
